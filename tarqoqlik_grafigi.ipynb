```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/merc.csv")
df.head()
```

```python
plt.figure(figsize=(10, 6)) # grafikka olcham berib oldik
sns.scatterplot(x="year", y="price", data=df) # year va price ortasidagi tarqoqlik aniqlab olish uchun scatterplot funksiani ishlatdik
plt.show()
```

```python
df = df[df.year>2000] # grafikga 2000 chi yildan katta bolgan qimatlarni chiqarishni buirdik
plt.figure(figsize=(10, 6)) # grafikga", da olcham berib oldik
sns.regplot(x="year", y="price", data=df) # "year" bilan price ortasidagi qimatlarni tarqoqligini korib oldik
plt.show()
```

```python
df = df[df.year>2000] # filtrlab oldik
plt.figure(figsize=(16, 8)) # grafikka olchov berib ham oldik
sns.scatterplot(x="mileage", y="price", hue='transmission', data=df) #sctatterplot funksiasi orqali 3 ustunni tarqoqligini korib ham oldik
plt.show()
```

```python
df = df[df.year>2000] # filtrlab oldiik
plt.figure(figsize=(20, 10)) # grafikka olchov berib oldik
sns.scatterplot(x="mileage", y="price", hue='fuelType', style='fuelType', s=100, data=df) # bu safar 4 ta ustuni qoshdik va fuelType ga style berib oldik
plt.show()
```

```python
sns.swarmplot(data=df, x=df['fuelType'], y=df['price']) # swarmplot orqali fuletpe bn price ni tarqoqligini korib oldik
plt.show()
```
