```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/merc.csv')
df=df[df.year>=2000] # 2010 - yildan kein chiqqan mashinalarni saqlab qoldik
df=df[df.transmission != 'Other'] # faqat mexanik, avtamot, yarim avtamot mashinalarni olib qolamiz
df=df[df.fuelType != 'Other'] # faqat benzin, gbrid, dizel mashinalarni olib qolamiz
df.head()
```

```python
plt.figure(figsize=(12, 6)) # grafikga olcham berib oldik
sns.histplot(x='price', data=df) # narh ustunini distribution qilib oldik
plt.show()
```

```python
sns.displot(data=df, x='price', hue='transmission', height=5, aspect=2) # displot price va transmissioni  disturbition qilib oldik va olcham ham berib oldik
plt.show()
```

```python
sns.displot(data=df, x='price', col='transmission', height=10, aspect=2) # tarnsmisssiondagi qimatlarni alohida 'col' funksiasi orqali garfik qilib chiqaeib oldik
plt.show()
```

taqsimot kde

```python
sns.displot(data=df, x='price', kind='kde', height=5, aspect=2) # kind=kde funksiasi orqali linia qilib oldik
plt.show()
```

```python
sns.displot(data=df, x='price', kde=True, height=5, aspect=2) # kde=True funksiasi orqali ham linia bn birga chiqarib oldik
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x='price', shade=True, color='red') # busafar kdeplot orqali sgade=True funksiasini qollagan holda boshqacha linia qilib odik
plt.show()
```

taqsimot cde

```python
sns.displot(data=df, x='price', kind='ecdf', height=5, aspect=2) # kind=hist
plt.xticks(rotation=90) # pasdagi yozuvlarni 90 katta qilib oldik
plt.grid(which='both') # grafikga reshotka berib oldik
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='mileage', kind='ecdf', height=5, aspect=2) # kind=hist
plt.xticks(rotation=90) # pasdagi yozuvlarni 90 gradusga katta qilib oldik
plt.grid(which='both') # grafikga reshotka berib oldik
plt.minorticks_on()
plt.show()
```
