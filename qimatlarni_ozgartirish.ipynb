map() qimatlani moslash

```python
kl = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data_nan.csv', index_col=0)
kl.head()
# githubdan jadval chaqirib olib uni dataframe korinishiga otkizib oldik va boshidan 5  qatorni chiqarib ham oldik
```

```python
kl['num-of-cylinders'].unique()
# num-of qimatlarni chiqarib oldik
```

```python
dict = {'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'eight':8, 'twelve':12}
dict # va dict orqali str qimatlarni int korinishiga otqizish uchun list korinishiga keltirib oldik
```

```python
kl['cylinders'] = kl['num-of-cylinders'].map(dict)
kl # va cylinders nomli yangi ustun berdik va num-of-cylinders gsa ulab .map(dict) qilib int larni kirgizib oldik
```

```python
kl['price']
# price ustunini chaqirib oldik
```

```python
narh = 12080
kl['prcie_uzb'] = kl['price'].map(lambda x:  x*100)
kl # yangi price_uzb ustun yaratib oldik va map orqali lambda x:  x*100 shu funksiani berib oldik
```

.replace() .rename() funksialri

```python
kl.replace(dict)
# .replace orqali istalgan ustunni qimatlarini ozgartirishimiz mumkun
```
