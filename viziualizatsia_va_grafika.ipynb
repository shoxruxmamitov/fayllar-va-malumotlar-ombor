```python
x = np.arange(0, 4*np.pi, 0.1)
t = np.sin(x) # x va t qimat berib oldik
```

```python
plt.plot(x, t, linewidth=3, linestyle='--', color='r')
plt.show()# x va t ni chaqirib oldik va qalinligini linewidth qilib 3 qalinligida da tenglashtirib oldik
          # va linstyle orqali style berib oldik color orqali esa grafikimizga rang berib oldik
```

```python
plt.figure(figsize=(10, 5)) # grafikga olchov berb oldik 10 ga 5 qilib
plt.plot(x, t, linewidth=2, linestyle='-', color='k', marker='o')
plt.title('sin (x) funksia') # girafikga nom berib oldik
plt.xlabel('pi') # x qimatga nom berib oldik
plt.grid(color='k', linestyle='--') # grafikga setka qoib oldik shu bn birga rang va shakl ham berib oldik
plt.show() #x va t ni chaqirib oldik va qalinligi, stil, rang va shakl berib oldik
```
