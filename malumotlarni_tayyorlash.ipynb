```python
import pandas as pd # pandas  ni chaqirib oldik
```

```python
import numpy as np # np ni chaqirib oldik
```

dropna funksiasi

```python
st = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data_nan.csv', index_col=0)
st # githubdan jadvalni dataframe qilib chaqirib oldik
```

```python
st.isnull() # mavjud bolmagan qimatlarni chaqirib oldik mavjud bolmagan qimatlarni True qilib belgilidi
```

```python
st.notnull() # notnull ham mavjud bolmagan qimatlarni chaqiradi va False qilib bwlgilidi
```

```python
st.isnull().sum() # qasi qatorda mavjud bolmagan qimatlarni sonini chiaqrib berdi
```

```python
len(st) # uzunligni hisoblab oldik
```

```python
st.isnull().sum()/len(st)*100 # mavjud bolmagan qimatlarni sonini chaqirib proporsiasini hisoblab va uning foizini 100 ga kopatirib bilib oldik
```

```python
st.dropna() # mavjud bolmagan qimatlarni ochrib oldik
```

```python
len(st.dropna()) # ochirib olgan qimatlarni sonini bilib oldik
```

```python
st.dropna(axis=1) #mavjud bolmagan ustunlarni tashlab oldik
```

```python
st.dropna(axis=1, thresh=48)
# mavjud bolmagan qimatlarni 48 tadan kam bolgan ustunni tashlashnu buirdik
```

```python
st.dropna(subset=['price'])
# price ustnini nan bolgan qimatlarni subset orqali ochirishni buirdik
```

.fillna() funksiasi

```python
st.sample(10)
```

```python
st.fillna(99)
# fillna orqali mavjud bolmagan qimatlarga 99 qimat berib oldik ln bu ozgincha xato
```

```python
st.fillna({'horsepower':87, 'whell-base':92})
#fillna orqali ozimiz hoxlagan ustunga qimat berib chiqdik
```

```python
st.fillna({'average-mileage':np.mean(st['average-mileage']), 'horsepower':np.mean(st['horsepower'])})
#average-mileage va horsepower qimatlariga ortancha qimatni berib oldik np.mean orqali
```

```python
st.fillna(method='ffill')
#ffill funksiasi orqali mavjud bolmagan qimatni ozidan bitta tepada bolgan qimatni oziga kochirib oladi
```

```python
st.fillna(method='bfill')
# bfill funksiasi orqali mavjud bolmagan qimat ozidan bitta past bolgan qimatdan oziga kochirib oladi
```

```python
st[st.company=='toyota'].fillna(method='bfill')
# company ustunidan toyota qimatiga bfill funksiasini berib oldik va qimat berdik
```
