```python
import pandas as pd
import sqlite3 as sql
# pandas va sqlite3 kutubxonasini chaqirib oldik
```

```python
db = sql.connect('newdb.db')
cursor = db.cursor()
# newdb bosh ombori hosil qilib oldik
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/Electricity_Production_By_Source.csv')
df# database ni kirgizib oldik
```

```python
cols = ['strana', 'code', 'ear', 'coal', 'gas', 'hdro', 'renewable', 'solar', 'oil', 'wind', 'nuclear']
df.columns = cols
df.head()
# usutnlar nomini ozimizga qula qilib almashtirib oldik va columns orqali qoshib oldik
```

```python
df.to_sql('elecit', db)
#tepadagi databasega 'elect' nomli db qoshib oldik
```

```python
cursor.execute('select name from sqlite_master where type="table"')
cursor.fetchall()
# data baseni ichida borligini tekshirib oldik
```

```python
cursor.execute('pragma table_info(elecit);')
cursor.fetchall()
# elecit nomli databaseni ichida qanaqa ustun borligi ni tekshirib oldik va qasi turligni ham
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/world_population.csv')
df.head()
# github dan world population nomli databazeni chaqirib oldik
```

```python
terik = """
create table if not exists korlik(
  rank int,
  countri text primary key,
  densit double,
  population double not null,
  area int not null
);
"""
cursor.execute(terik)
db.commit()
# korlik nomli data hosil qilib oldik
```

```python
uz = df[df.country == 'Uzbekistan']
uz
# df jadvaldan uzbekistanga tegishli bolgan malmotlarni chiqarib oldik
```

```python
rank = uz.index.values[0]
country = uz.at[rank,'country']
density	= uz.at[rank, 'density']
population = uz.at[rank, 'pop2021']
area = uz.at[rank, 'area']
print(f'{rank}, {country}, {density}, {population}, {area}')
# malumotlarni alohida massiv qilib chiqarib oldik
```

```python
cueer = f"""
insert into korlik values({rank}, '{country}', {density}, {population}, {area})
"""
cursor.execute(cueer)
db.commit()
# korlik databazega shu malumlarni kirgizb qodik
```

```python
cursor.execute('select * from korlik')
cursor.fetchall()
#korlik databazeni malumotlarni ciqarib oldik
```
