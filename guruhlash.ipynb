```python
bins = [0, 1000, 10000, 30000, 50000, 100000, 300000, 1500000]
# list korinishida sonlarni hosil qildik
```

```python
popilation = tr.pop2021
```

```python
guruh = pd.cut(popilation, bins)
guruh # guruhlarga bolib oldik oraliqlarni bilib oldish uchun ishlatdik
```

```python
guruh.value_counts()
# guruhlagan sonlarimizni nechtaligi ni chiqarib oldik
```

```python
tr['guruh'] = guruh
tr.head()
# guruh nomli ustunni qoshib oldik
```

```python
lk = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data.csv', index_col=0)
lk.sample(7)
```

```python
lk.price.describe()
```

```python
binss = [0, 10000, 20000, 50000]
```

```python
range = pd.cut(lk.price, binss)
range #price narhiga binss ni boglab qodik sonlar oriligini korsatish uchun
```

```python
range.value_counts()
```

```python
names = ['arzon', "o'rta", 'qimmat']
range = pd.cut(lk.price, binss, labels=names)
range.value_counts()
```

```python
ot = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data.csv', index_col=0)
ot.tail() # dataframeni chaqirib oldik
```

```python
ot.groupby('company').price.max() #compani ni eng qimmat narhini bilib oldik
```

```python
ot.groupby('company').price.mean() # compan ni ortancha narhini bilib oldik
```
