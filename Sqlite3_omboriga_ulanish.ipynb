```python
db = sql.connect('chinook.db')
cursor = db.cursor()
# database ni saqlab olib cursor nomli ozgaruvchiga chaqirib oldik
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
# chinook zip failini ichida nechta database borligini aniqlab oldik
```

```python
cursor.execute('select count (*) from customers')
print(cursor.fetchone())
# chinook zip failini ichidagi 'customers' databse ni ichidan nechta qimat borligini bilib oldik
```

```python
cursor.execute('select count (*) from albums')
cursor.fetchall() # bunda ham qimatlarni korib oldik
```

```python
cursor.execute('select count (*) from artists')
cursor.fetchmany()
```

```python
cursor.execute ('select count (*) from tracks')
cursor.fetchone()
```

```python
cursor.execute('select count(*) from invoices')
cursor.fetchone()
```

Qimatlarni chiqarish

```python
cursor.execute('pragma table_info(customers);')
cursor.fetchall()
# customers database ni qimatlarini chiqarib oldik
```

```python
cursor.execute('pragma table_info(albums);')
cursor.fetchone()
# albums databaseni ustunlarini chiqarib oldik
```

```python
cursor.execute('pragma table_info (artists);')
cursor.fetchmany()
```
