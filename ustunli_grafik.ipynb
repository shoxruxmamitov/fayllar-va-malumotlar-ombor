```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# kutubxonalarni chaqirib oldik
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/uzbekistan.csv', index_col=0)
df.columns = ['maydoni', 'aholi']
df # dataframeni chaqirib olib columnns ga uzimiz nom berib oldik
```

```python
plt.figure(figsize=(20, 10)) # olcham berib oldik
sns.barplot(x=df.index, y=df['aholi']) #

plt.title("O'zbekiston aholisi") # grafikga nom berib oldik
plt.xlabel("Tumanlar") # x qimatga ham nom berdik
plt.ylabel("Aholi soni") # y qimatga ham nom berdik
plt.grid() # chizmaga gorizantal va vertikal chiziq qoshib oldik
plt.xticks(rotation=90) # grafik pasidagi yozuvlarni orasini katta qilib olyik
plt.show()
```

HEATMAP funksiasi

```python
lm = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/museum_visitors.csv", index_col=0, parse_dates=[0], usecols=[0, 1])
lm.head()
```

```python
type(lm.index)
```

```python
lm['month'] = [i.month for i in lm.index] # date columdan month ni ajratib oldik
lm['year'] = [i.year for i in lm.index] # date columdan yearearni ajratib olyearik
lm
```

```python
lm2 = lm.groupby(['month', 'year']).max()
lm2 # lm yatframni alohiya guppaga taqsimlab oldik
```

```python
lm2.unstack(level=0)
#tepadagi funksianing teskarisi
```

```python
plt.figure(figsize=(12, 6)) # grafikga razmer berib oldik
sns.heatmap(data=lm2.unstack(level=0), cmap='YlGnBu', linewidths=1, square=True) # sns kutubxonasini chaqirib oldik rang berib oldik
plt.show()
```
