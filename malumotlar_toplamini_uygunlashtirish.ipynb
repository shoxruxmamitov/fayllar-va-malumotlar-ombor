sodda misollar

```python
A = pd.DataFrame({'Transport': ['poezd', 'avtomobil', 'kema'],
                  'Tezligi': ['70', '185', '65']})
A # dataframe hosil qilib olamiz
```

```python
B = pd.DataFrame({'Transport': ['poezd', 'avtomobil', 'tramvay'],
                  'Narhi':['98000', '110000', '132000']})
B # ikkinchi dataframe hpsil qilib oldik bunda 2 la df da Transport ustuni bir hil
```

merge funksiasi

```python
merged = pd.merge(A, B)
merged # ikkla adataframeni merge funksiasi orqali qoshib oldik va bunda bizaga tezlik va narh bitta dataframeda boldi ln tramvay va kema qoshilmadi
# bu funksia INNER JOIN (ichki qoshilish) deb ataladi
```

```python
merged = pd.merge(A, B, how='outer')
merged # bu funksia da 2 ta dataframeda mavjud bolgan va mavjud bolmagan dataframe qimatlari qoshilavi qoshilganda bir qimat b9lmassa nan qimat beradi
 # bu bizada OUTER JOIN (tashqi qoshilish) deb ataladi
```

```python
merged_lift = pd.merge(A, B, how='left') # how = 'left' (chap tomon qoshilish) deb ataladi
merged_lift # bu funksia 2 la dataframedan ustunlarni olib ln (how=left) qilganligimiz uchun faqat A datafrmedan hamma indexlarni oldi
            # ln B dataframedan 2 la dataframe bir xil bolmagan qimatlarni olmaydi
# how = 'left' funksiasi bizga faqat ongdagi dataframeni ustunlarini 100 foiz qilib chiqarib beradi
```

```python
merged_right = pd.merge(A, B, how='right') # how = 'right' (ong tomon qoshilish) deb t=ataladi
merged_right #  # bu funksia 2 la dataframedan ustunlarni olib ln (how=right) qilganligimiz uchun faqat B datafrmedan hamma indexlarni oldi
            # ln A dataframedan 2 la dataframe bir xil bolmagan qimatlarni olmaydi
# how = 'right' funksiasi bizga faqat chapdagi dataframeni ustunlarini 100 foiz qilib chiqarib beradi
```

df

```python
R = pd.DataFrame({'Transport': ['Avtomobil', 'Kema', 'Belaz'],
                  "Tezligi": ['220', '230', '150']})
R # dataframe hosil qilib oldik
```

```python
S = pd.DataFrame({'Transport_turi': ['Avtomobil', 'Kema', 'Tank'],
                  "Og'irligi": ['23', '178', '78']})
S #dataframe hosil qilib oldik
```

```python
merg = pd.merge(R, S, left_on='Transport', right_on='Transport_turi')
merg # bu funksia orqali ikkala dataframeni qoshib oldik left va right tomondan ham ustunlarni qoshib oldik
```

```python
merg1 = merg[['Transport', 'Tezligi', "Og'irligi"]]
merg1 # merg dan ozimizga kk bolgan ustunlarni qoldirdik
```

agar bizda indekslar har xil bolsa

```python
merged1 = pd.merge(S, R, left_index=True, right_on='Transport')
merged1 # bu funksiya da S da indeximiz bolmagan ligimi sababli left_index=True qildik
```
