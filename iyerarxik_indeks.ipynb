```python
df = pd.read_csv('/content/Practicum_Datasets/world_cities_air_water_quality.csv')
df.head() #praktikum datasets ichidan ozimizga kk li bolan failni chaqirib olamiz
```

```python
df.info() # dataframe haqida malumotlarni koramiz
```

```python
columns = ['City', 'Region', 'Country', 'AirQuality', 'WaterPollution']
df.columns = columns
df.head()
#bizda columnlari qoshitirnoq ichida bolganligi uchun biz uni qoshtirnoqsiz qilib qodik sababi bizga muammo iolib kelishi mumkun va
#dataframeda hunuk korinadi
```

```python
df['Region'] = df['Region'].map(lambda x:x.replace('"', ''))
df
#biz region columining qimatida qoshtirnoqlarni olib tashladik
```

```python
df['Country'] = df['Country'].map(lambda x:x.replace('"', ''))
df
# bu safar regio coumining qimatlardagi qoshtinoqlarni olib tashladik
```

```python
df.to_csv('world_cities_air_water_quality.csv', index=False)
# biz bu dataframeni saqlab oldik
```

```python
df[df['Country']=='Uzbekistan']
# Country ustunidan Uzbekistonga tegishli bolgan malumotlarni chaqirib oldik ln hech narsa chiqmadi
# sababi ' Uzbekiston' lugatida qoshtirnogida joy tashlab ketilgan bu ham bniza noqulaylik keltiradi
```

```python
df['Country'] = df['Country'].map(lambda x:x.strip())
# biz bu muamoni tugirlab oldik lambda va strip orqali v.h
```

```python
df['Country'][3] # tugirlanga tugirlanmaganini tekshirib oldik
```

iyerarxik malumotlar ustuni

```python
df2 = df1.copy()
df2.sample(7)
# df1 datframeni df2 ga copi qilib kochirib oldik va 5 ta qatorni chaqirib oldik
```

```python
df2['AirPollution'] = 100-df2['AirQuality']
df2 #airquality ustuniga teskari bilgan yangi ustun hosil qildik
```

```python
df2['WaterQuality'] = 100-df2['WaterPollution']
df2 #waterpollution ga teskari bolgan waterqualt ustunini hosil qilib oldik
```

```python
df3 = pd.DataFrame(df2.values, index=df2.index, columns=[['AIR', 'AIR', 'WATER', 'WATER'], ['AirQuality', 'AirPollution',	'WaterPollution', 'WaterQuality']])
df3
```

```python
df3['AIR'] #air ustunini chaqirib oldik
```

```python
df3['WATER']# WATER ga tegishli bolgan ustunni chaqirib oldik
```

```python
df3.columns.names = ['Object', 'Info']
df3 # ustunlarnigng oziga ham alohida nom berib oldik
```
