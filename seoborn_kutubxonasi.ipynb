```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# kutubxonalarni chaqirib oldk
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/museum_visitors.csv',index_col=0)
df.head() # dataframeni chaqirib oldik
```

```python
sns.lineplot(data=df)
plt.show() # sns.line plot orqali dataframeni grafik korinishiga keltirib oldik
```

```python
plt.figure(figsize=(12,7)) # grafigimizga olcham berib oldik 12 ga 7 qilib
sns.lineplot(data=df) # sns.lineplot orqali df ni data ga tenglashtirib oldik
plt.xlabel('sana') # x qiymatga sana nomini berib oldik
plt.ylabel('odamlar soni') # y qiymatga ham nom berib oldik
plt.xticks(rotation=90) # sanalarini orasini katta qilib qo'dik
plt.show()
```
